Separating the features and target

```python
X=data_frame.drop(columns=['label'],axis=1)
Y=data_frame['label']
```

**Splitting the data into training data and testing data**


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
scaler=StandardScaler()
# This is the scaled data for the breast cancer diagnosis dataset (30 features)
# Ensure we are using the X_train and Y_train from the diagnosis dataset
# These variables were defined earlier in the notebook
X_train_cancer_std=scaler.fit_transform(X_train) # X_train here should be the diagnosis training data
X_test_cancer_std=scaler.transform(X_test) # X_test here should be the diagnosis testing data
```

**Building a Neural Network**

```python
#importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of Neural Network
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')
])
```

```python
#compiling the Neural Network
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#training the Neural network
# Use the scaled training data for the breast cancer diagnosis dataset
history=model.fit(X_train_cancer_std,Y_train,validation_split=0.1,epochs=10)
```

**Visualizing accuracy and loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

# ** which stage of breast cancer **

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
df = pd.read_csv('/content/sample_data/breast_cancer_stage_prediction_dataset_500.csv')
```

```python
#number of rows and columns in stage predicition dataset
df.shape
```

```python
print(df)
```

```python
df.info()
```

```python
df.shape
```

```python
df['Stage'].value_counts()
```

```python
# Encode categorical columns before splitting
categorical_cols = ['T_Stage', 'N_Stage', 'M_Stage', 'ER_Status', 'PR_Status', 'HER2_Status', 'Stage']
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Features and Labels
X = df.drop("Stage", axis=1)
y = df["Stage"]
```

```python
# One-hot encoding for categorical features (excluding label)
X = pd.get_dummies(X, columns=['T_Stage', 'N_Stage', 'M_Stage', 'ER_Status', 'PR_Status', 'HER2_Status'])
```

```python
# Apply SMOTE to balance classes
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

print("\nResampled class distribution:")
print(Counter(y_resampled))
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=2)
```

```python
# Model
input_dim = X_train.shape[1]
num_classes = len(np.unique(y))

model_stage = keras.Sequential([
    keras.layers.Dense(40, activation='relu', input_shape=(input_dim,)),
    keras.layers.Dense(num_classes, activation='softmax')
])

model_stage.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train model
history = model_stage.fit(X_train, y_train, validation_split=0.1, epochs=10)
```

```python
# Plot accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Stage Prediction Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Plot loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Stage Prediction Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Evaluate
loss, acc = model_stage.evaluate(X_test, y_test)
print(f"✅ Test Accuracy for Stage Prediction: {acc:.2f}")
```

```python
# Predict for new input
input_data_stage = [10, 20, 56, 'T2', 'N1', 'M0', 'Positive', 'Positive', 'Negative', 60]
categorical_input = ['T_Stage', 'N_Stage', 'M_Stage', 'ER_Status', 'PR_Status', 'HER2_Status']

# Label encode categorical values
for i, col in enumerate(categorical_input):
    input_data_stage[3 + i] = label_encoders[col].transform([input_data_stage[3 + i]])[0]

input_df = pd.DataFrame([input_data_stage], columns=X.columns[:10])
input_df = pd.get_dummies(input_df)

# Align input with training data
input_df = input_df.reindex(columns=X.columns, fill_value=0)

# Scale
input_std = scaler.transform(input_df)

# Predict
prediction = model_stage.predict(input_std)
predicted_index = np.argmax(prediction)
predicted_stage = label_encoders['Stage'].inverse_transform([predicted_index])[0]

print(f"🩺 Predicted Cancer Stage: {predicted_stage}")
```
